```python
import geopandas
parcels = geopandas.read_file("Edina_Parcels.shp")
parcels.head()
```

```python
#parcels.explore(column="ShapeSTAre", scheme="percentiles", tiles="CartoDB positron", legend=True)
```

```python
parcels.explore(column="ShapeSTAre", scheme="percentiles", tiles="CartoDB positron", legend=True).save("Edina_Parcels.html")
```

```python
#filtering for residential properties and excluding schools, churches via null sale_code
residential = parcels[(parcels["PR_TYP_NM1"] == 'RESIDENTIAL') & (parcels["SALE_CODE"].notnull())]
residential.head()
```

```python
#residential.explore(column="ShapeSTAre", scheme="percentiles", tiles="CartoDB positron", legend=True)
```

```python

residential.explore(column="ShapeSTAre", scheme="percentiles", tiles="CartoDB positron", legend=True).save("Edina_Residential_Parcels.html")
```
